```python
import sys
sys.path.append('../..')

import requests
import json
from dev.utils import load_config

API_KEY = load_config()["circle"]["api_key"]
HEADERS = {
    'Content-Type': 'application/json',
    'Authorization': f'Bearer {API_KEY}'
}
```

```python

URL_ME = "https://app.circle.so/api/v1/me"
URL_COMMUNITY = "https://app.circle.so/api/v1/communities"
URL_SPACES = "https://app.circle.so/api/v1/spaces"
URL_POSTS = "https://app.circle.so/api/v1/posts"
URL_COURS_LESSONS = "https://app.circle.so/api/v1/course_lessons"
URL_COURS_SECTIONS = "https://app.circle.so/api/v1/course_sections"

def call_spaces(community_id=52148):
    url = f'{URL_SPACES}?community_id={community_id}'
    response = requests.get(url, headers=HEADERS, data=json.dumps({}))
    if response.status_code != 200:
        raise ValueError
    return response.json()

def call_posts(community_id=52148, space_id=471612):
    url = f'{URL_POSTS}?community_id={community_id}&space_id={space_id}'
    response = requests.get(url, headers=HEADERS, data=json.dumps({}))
    if response.status_code != 200:
        raise ValueError
    return response.json()

def call_course_lessons(community_id=52148):
    url = f'{URL_COURS_LESSONS}?community_id={community_id}'
    response = requests.get(url, headers=HEADERS, data=json.dumps({}))
    if response.status_code != 200:
        raise ValueError
    return response.json()

def call_course_lessons(community_id=52148):
    url = f'{URL_COURS_SECTIONS}?community_id={community_id}'
    response = requests.get(url, headers=HEADERS, data=json.dumps({}))
    if response.status_code != 200:
        raise ValueError
    return response.json()
```
