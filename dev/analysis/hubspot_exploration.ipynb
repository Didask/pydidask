## look for gclid in all columns of all tables of google ads

```python
table_schema = 'GOOGLE_ADS_AIRBYTE'
engine_base = create_snowflake_engine(database="RAW", schema="INFORMATION_SCHEMA")
sql_query = f"""
select table_name
from tables
where 
    table_schema = '{table_schema}'
order by table_name
"""
```

```python
google_ads_table_names = pd.read_sql_query(sql_query, engine_base)["table_name"]
```

```python
table_columns = dict()
for table_name in google_ads_table_names:
    query = f"""
    select column_name 
    from columns
    where 
        table_name = '{table_name}'
        and table_schema = '{table_schema}'
    order by ordinal_position
    """
    table_columns[table_name] = pd.read_sql_query(query, engine_base)["column_name"]
```

```python
all_columns = list()
for e in list(table_columns.values()):
    all_columns += [a.lower() for a in list(e.values)]
```

```python
[c for c in all_columns if "conversion" in c]
```
